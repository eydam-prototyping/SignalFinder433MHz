# Princeton 24bit

```python
import json
from matplotlib import pyplot as plt

raw_data = []
with open('sig_princeton24bit_sample.json', 'r') as f:
    raw_data = json.load(f)

print(len(raw_data))
```

```python
x = [p[0] for p in raw_data]
y = [p[1] for p in raw_data]

fig, ax = plt.subplots()
ax.scatter(x,y, s=.01)
fig.set_size_inches(5, 5)
ax.set_xlim(0, 1500)
ax.set_ylim(0, 1500)
ax.grid(True)
```
